```python
#Add Details: Input and store new records, including athlete information, event details, country,
#and performance metrics.
data=dict()
def add_details():
    print("Welcome to The Great Britain Olympics Input Dashboard ")
    print()
    t=int(input("Enter the number of Games you are interested in: "))
    games=[]
    
    for i in range(t):
        sports_name=input("Enter sports name: ")
        print()
        games.append(sports_name)
    years=input("Enter the years of olympics with spaces between them( for eg. '2004 2008 2012')")
    years=years.split(" ")
    print(games)
    print(years)
    
    for year in years:
        print(f"********** Entering Details for {year} Olympics**********")
        print()
        game_details=dict()
        for game in games:
            players_achieve=[]
            
            nos_participants=int(input(f"Enter number of participants from Great Britain for  {game} Olympics {year} :"))
            print()
            
            for i in range(nos_participants):
                player_name=input(f"Enter player {i+1} name:")
                medal=input(f"Did {player_name}  win Medal in  {game} Olympics {year} (Enter y/n) :")
                print()
                
                if medal=="y":
                    gsb=input("Did he win 'gold'/'silver'/'bronze':  ")
                    print()
                else:
                    gsb=0
                    
                players_achieve.append((player_name,gsb))
            if players_achieve:
                game_details[game]=players_achieve
        if game_details:
            data[year]=game_details
    return   
      
    
#add_details()
    
    
```

```python
def view_details(data):
    print("Welcome to The Great Britain Olympics View Dashboard")
    print("Available Olympic Years:")
    
    # Display available years
    years = list(data.keys())
    for i, year in enumerate(years):
        print(f"{i + 1}. {year}")
    
    # Prompt user to choose a valid year
    year_choice = int(input("Choose a year by entering the corresponding number: ")) - 1
    
    if year_choice < 0 or year_choice >= len(years):
        print("Invalid choice. Please restart and choose a valid year.")
        return
    
    chosen_year = years[year_choice]
    
    print(f"\nAvailable sports in the {chosen_year} Olympics:")
    
    # Display available sports for the chosen year
    sports = list(data[chosen_year].keys())
    for i, sport in enumerate(sports):
        print(f"{i + 1}. {sport}")
    
    # Prompt user to choose a valid sport
    sport_choice = int(input("Choose a sport by entering the corresponding number: ")) - 1
    
    if sport_choice < 0 or sport_choice >= len(sports):
        print("Invalid choice. Please restart and choose a valid sport.")
        return
    
    chosen_sport = sports[sport_choice]
    
    print(f"\nParticipants in {chosen_sport} at the {chosen_year} Olympics:")
    
    # Display participants for the chosen sport
    participants = data[chosen_year][chosen_sport]
    for i, (participant, _) in enumerate(participants):
        print(f"{i + 1}. {participant}")
    
    # Prompt user to choose a valid participant
    participant_choice = int(input("Choose a participant by entering the corresponding number: ")) - 1
    
    if participant_choice < 0 or participant_choice >= len(participants):
        print("Invalid choice. Please restart and choose a valid participant.")
        return
    
    chosen_participant, achievement = participants[participant_choice]
    
    # Count the number of medals
    medal_count = {
        'gold': 0,
        'silver': 0,
        'bronze': 0
    }
    
    if achievement in medal_count:
        medal_count[achievement] += 1
    
    total_medals = sum(medal_count.values())
    
    print(f"\nDetails for {chosen_participant} in {chosen_sport} at the {chosen_year} Olympics:")
    print(f"Gold Medals: {medal_count['gold']}")
    print(f"Silver Medals: {medal_count['silver']}")
    print(f"Bronze Medals: {medal_count['bronze']}")
    print(f"Total Medals: {total_medals}")
    print("--------------------------------------------")
    print("End of details.\n")

# Example usage:
#view_details(data)
```

```python
def view_medal_count(data):
    # Initialize counters for medals across all years
    total_gold_all_years = 0
    total_silver_all_years = 0
    total_bronze_all_years = 0

    # Retrieving sports played, medals won, and calculating totals
    for year, sports in data.items():
        # Initialize counters for medals for the current year
        total_gold_year = 0
        total_silver_year = 0
        total_bronze_year = 0

        print(f"Sports played in {year}:")
        for sport, participants in sports.items():
            print(f"- {sport}")
            for participant, medal in participants:
                if medal == 'gold':
                    total_gold_year += 1
                elif medal == 'silver':
                    total_silver_year += 1
                elif medal == 'bronze':
                    total_bronze_year += 1

        # Print the total medals for the current year
        print(f"Total medals won in {year}:")
        print(f"  Gold: {total_gold_year}")
        print(f"  Silver: {total_silver_year}")
        print(f"  Bronze: {total_bronze_year}")
        print(f"  Total: {total_gold_year+total_silver_year+total_bronze_year}")
        print()  # For better readability

        # Add the year's totals to the overall totals
        total_gold_all_years += total_gold_year
        total_silver_all_years += total_silver_year
        total_bronze_all_years += total_bronze_year

    # Print the total medals across all years
    print(f"Total medals won in all years together:")
    print(f"  Gold: {total_gold_all_years}")
    print(f"  Silver: {total_silver_all_years}")
    print(f"  Bronze: {total_bronze_all_years}")
    
    
#view_medal_count(data)
```

```python
def search_athlete_performance(data):
    # Collect all unique athlete names
    athletes = set()
    for sports in data.values():
        for participants in sports.values():
            for participant, _ in participants:
                athletes.add(participant)
    
    # Convert the set to a sorted list for display
    athletes_list = sorted(athletes)

    # Display the list of players
    print("List of athletes:")
    for idx, athlete in enumerate(athletes_list, 1):
        print(f"{idx}. {athlete}")

    # Prompt the user to select an athlete
    try:
        choice = int(input("Enter the number of the athlete you want to search for: "))
        if 1 <= choice <= len(athletes_list):
            athlete_name = athletes_list[choice - 1]
        else:
            print("Invalid choice. Please enter a number from the list.")
            return
    except ValueError:
        print("Invalid input. Please enter a number.")
        return

    # Initialize counters for the athlete's medals
    total_gold = 0
    total_silver = 0
    total_bronze = 0

    # Initialize a list to store performance details
    performance_details = []

    # Search through the data for the specified athlete
    for year, sports in data.items():
        for sport, participants in sports.items():
            for participant, medal in participants:
                if participant == athlete_name:
                    # Update medal counters based on the type of medal
                    if medal == 'gold':
                        total_gold += 1
                    elif medal == 'silver':
                        total_silver += 1
                    elif medal == 'bronze':
                        total_bronze += 1

                    # Add the performance detail to the list
                    performance_details.append({
                        'year': year,
                        'sport': sport,
                        'medal': medal
                    })

    # Display the athlete's performance
    if performance_details:
        print(f"\nPerformance statistics for {athlete_name}:")
        for detail in performance_details:
            medal_display = detail['medal'].capitalize() if detail['medal'] != 0 else 'No medal'
            print(f"  Year: {detail['year']}, Sport: {detail['sport']}, Medal: {medal_display}")

        # Print total medals won by the athlete
        print(f"\nTotal medals won by {athlete_name}:")
        print(f"  Gold: {total_gold}")
        print(f"  Silver: {total_silver}")
        print(f"  Bronze: {total_bronze}")
        print(f"  Total: {total_gold + total_silver + total_bronze}")
    else:
        print(f"No performance data found for {athlete_name}.")
        
#search_athlete_performance(data)
```

```python
def view_event_details(data):
    print("Welcome to The Great Britain Olympics Event Details Dashboard")
    
    # Display available years
    print("\nAvailable Olympic Years:")
    years = list(data.keys())
    for i, year in enumerate(years):
        print(f"{i + 1}. {year}")
    
    # Prompt user to choose a valid year
    year_choice = int(input("Choose a year by entering the corresponding number: ")) - 1
    
    if year_choice < 0 or year_choice >= len(years):
        print("Invalid choice. Please restart and choose a valid year.")
        return
    
    chosen_year = years[year_choice]
    
    # Display available sports for the chosen year
    print(f"\nAvailable sports in the {chosen_year} Olympics:")
    sports = list(data[chosen_year].keys())
    for i, sport in enumerate(sports):
        print(f"{i + 1}. {sport}")
    
    # Prompt user to choose a valid sport
    sport_choice = int(input("Choose a sport by entering the corresponding number: ")) - 1
    
    if sport_choice < 0 or sport_choice >= len(sports):
        print("Invalid choice. Please restart and choose a valid sport.")
        return
    
    chosen_sport = sports[sport_choice]
    
    print(f"\nDetails for {chosen_sport} in the {chosen_year} Olympics:")
    print("--------------------------------------------")
    
    # Display participants for the chosen sport
    participants = data[chosen_year][chosen_sport]
    if not participants:
        print(f"No participants data available for {chosen_sport} in the {chosen_year} Olympics.")
        return
    
    for participant, achievement in participants:
        if achievement == 0:
            print(f"{participant} participated but did not win a medal.")
        else:
            print(f"{participant} won a {achievement} medal.")
    
    print("--------------------------------------------")
    print("End of event details.\n")

# Example usage:
#view_event_details(data)
```

```python
def view_statistics(data):
    print("Welcome to The Great Britain Olympics Statistics Dashboard")
    
    # Initialize counters for statistics
    total_events = 0
    total_participants = 0
    total_gold = 0
    total_silver = 0
    total_bronze = 0
    
    # Iterate through the data to calculate statistics
    for year, sports in data.items():
        for sport, participants in sports.items():
            total_events += 1
            total_participants += len(participants)
            
            for participant, achievement in participants:
                if achievement == 'gold':
                    total_gold += 1
                elif achievement == 'silver':
                    total_silver += 1
                elif achievement == 'bronze':
                    total_bronze += 1
    
    # Calculate averages
    avg_participants_per_event = total_participants / total_events if total_events > 0 else 0
    avg_gold_per_event = total_gold / total_events if total_events > 0 else 0
    avg_silver_per_event = total_silver / total_events if total_events > 0 else 0
    avg_bronze_per_event = total_bronze / total_events if total_events > 0 else 0
    
    # Medal distribution
    total_medals = total_gold + total_silver + total_bronze
    gold_distribution = (total_gold / total_medals) * 100 if total_medals > 0 else 0
    silver_distribution = (total_silver / total_medals) * 100 if total_medals > 0 else 0
    bronze_distribution = (total_bronze / total_medals) * 100 if total_medals > 0 else 0
    
    # Display statistics
    print("\nStatistics Summary:")
    print(f"Total Events: {total_events}")
    print(f"Total Participants: {total_participants}")
    print(f"Average Participants per Event: {avg_participants_per_event:.2f}")
    print(f"Total Gold Medals: {total_gold}")
    print(f"Total Silver Medals: {total_silver}")
    print(f"Total Bronze Medals: {total_bronze}")
    print(f"Average Gold Medals per Event: {avg_gold_per_event:.2f}")
    print(f"Average Silver Medals per Event: {avg_silver_per_event:.2f}")
    print(f"Average Bronze Medals per Event: {avg_bronze_per_event:.2f}")
    print("\nMedal Distribution:")
    print(f"Gold Medals: {gold_distribution:.2f}%")
    print(f"Silver Medals: {silver_distribution:.2f}%")
    print(f"Bronze Medals: {bronze_distribution:.2f}%")
    print("--------------------------------------------")
    print("End of statistics overview.\n")

# Example usage:
#view_statistics(data)

```
